## 1. Suffix Tree Indexing

To achieve 10GB/min search speeds, linear scans are insufficient. We implement an optimized Suffix Array or Suffix Tree to allow $O(m)$ lookup for patterns of length $m$.

```python
def build_suffix_array(text):
    # SA-IS algorithm implementation for O(n) construction
    suffixes = sorted(range(len(text)), key=lambda i: text[i:])
    return suffixes

print("Suffix Array builder initialized.")
```
